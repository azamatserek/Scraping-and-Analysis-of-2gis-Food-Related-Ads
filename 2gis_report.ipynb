```python
# Importing the scraped data and showing scatter plot of search results position to the average check of restraurants Almaty 
```

```python
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_excel('output.xlsx',header=0,index_col=0)
plt.plot(df.checks)
plt.xlabel('Search Results Position')
plt.ylabel('Average Check (tenge)')
plt.title('Average Check per Search Results Position (0 -- means undefined average check)')
plt.savefig('searchResultsPosition_averageCheck.png')
plt.show()

```

```python
# Importing the scraped data and showing histogram  of average check of restraurants Almaty 
```

```python
plt.hist(df.checks, facecolor='g', alpha=0.75)
plt.xlabel('Average check (tenge)')
plt.ylabel('Frequency')
plt.title('Average Check Histogram (0 -- means undefined average check)')
plt.savefig('histogram_averageCheck.png')
plt.show()
```
